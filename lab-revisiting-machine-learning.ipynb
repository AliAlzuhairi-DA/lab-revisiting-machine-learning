```python
null_values = df.isnull().sum()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_null_values = null_values[categorical_columns]
```

```python
print("Null values in categorical columns:")
print(categorical_null_values)
```

```python
drop_list = []
```

```python
drop_list.extend(['OSOURCE', 'ZIP'])
print("Columns to drop:", drop_list)
```

```python
categorical_missing_percentages = df[categorical_columns].isnull().mean() * 100
drop_list.extend(categorical_missing_percentages[categorical_missing_percentages > 85].index.tolist())
print("Categorical columns with over 85% missing values:", drop_list)
```

```python
df.drop(columns=drop_list, inplace=True)
```
